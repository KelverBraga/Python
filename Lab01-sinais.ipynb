## 1) Gerar os seguintes sinais e visualiza-los (comandos plot e stem). 

```python
from pylab import *
```

### a) impulso unitário, para valores de n (abscissa) entre -10 e 10 (stem).

```python
n = np.arange(-10,11,1)
impulso = zeros(len(n))
impulso[int(len(impulso)/2)] = 1
matplotlib.pyplot.stem(n, impulso)
```

### b) degrau unitário, para n entre -10 e 10 (stem).


```python
n = np.arange(-10,11,1)
degrau = zeros(len(n))
metade = int(len(n) - len(n)/2)
print(metade)
degrau[metade:] = ones(metade+1)
matplotlib.pyplot.stem(n, degrau)
```

### c) ruido branco gaussiano (média zero, variância unitária) para n=0 a 40 (stem).

```python
n = np.arange(0,40,1)
ruido = np.random.randn(len(n))
matplotlib.pyplot.stem(n, ruido)
```

### d) senoide com 4 ciclos, para n=0 a 90 (stem).

```python
n = np.arange(0,90,1)
seno = np.sin(np.radians(n)*4*360/90)
matplotlib.pyplot.stem(n, seno)
```

### e) cosenoide com 5 ciclos, para n=0 a 110 (stem). 

```python
n = np.arange(0,110,1)
cosseno = np.cos(np.radians(n)*5*360/110)
matplotlib.pyplot.stem(n, cosseno)
```

### f) exponencial complexa com 5 ciclos, para n=0 a 70. Parte real (stem) e parte imaginária (plot). 

```python
n = np.arange(0,70,1)
funcao = exp(1j*n*164/360)
subplot(2, 1, 1)
matplotlib.pyplot.stem(n, real(funcao))
subplot(2, 1, 2)
matplotlib.pyplot.plot(n, imag(funcao))
```

### g) senoide com frequência w=0,2 rd ( sin(w*n) ) , para n de 0 a 100 (plot).

```python
n = np.arange(0, 100, 1)
w = 0.2
seno = np.sin(w*n)
matplotlib.pyplot.plot(n, seno)
```

### h) senoide de 60 Hz, (amostragem = 800 amostras/s), para n de 0 a 70 (plot).

```python
n = np.arange(0, 70, 1.0275)
w = 2*np.pi*60
seno = np.sin(w*np.radians(n))
matplotlib.pyplot.plot(seno)
```

### l) exponencial decrescente para n = 0 a 60 (stem).

```python
n = np.arange(0, 60, 1)
funcao = exp(-0.1*n)
matplotlib.pyplot.stem(n, funcao)
```

### m) exponencial oscilante decrescente, para n=0 a 90, parte real (stem).

```python
n = np.arange(0, 90, 1)
z = 0.97*exp(1j*0.31)
matplotlib.pyplot.stem(real(z**n))
```

### n) exponencial oscilante decrescente, para n=0 a 80, parte imaginaria (stem).

```python
n = np.arange(0, 80, 1)
z = 1.03*exp(1j*0.30)
matplotlib.pyplot.stem(imag(z**n))
```
