1) Considere a sequência teórica: $x[n]=1$, para $n=0,1,2,3$, e $x[n]=0$, para todos os outros valores de $n$ ($-\infty ... +\infty)$.

2) Calcule a transformada de fourier da sequencia $x[n]$, $n=0$ a 199, obtendo 200 valores na faixa de frequência entre 0 e $2\pi$.

```python
w = linspace(0, 2*pi, 200, endpoint=False)
x = zeros(200)
x[:4] = ones(4)
figure(1)
stem(x)
figure(2)
plot(w, abs(fft(x)))
```

3) Calcule a DFT desta sequencia, para n=0 a 3.

```python
w1 = linspace(0, 2*pi, 4, endpoint=False)
plot(w1, abs(fft(x[:4])))
```

4) Calcule a DFT desta sequencia, para n=0 a 7.

```python
w2 = linspace(0, 2*pi, 8, endpoint=False)
plot(w2, abs(fft(x[:8])))
```

5) Calcule a DFT desta sequencia, para n=0 a 15.

```python
w3 = linspace(0, 2*pi, 16, endpoint=False)
plot(w3, abs(fft(x[:16])))
```

6) Plote no mesmo eixo (abscissa de zero a 2$\pi$), todas as 4 transformadas. usando os seguintes simbolos: 2) linha continua, 3) 'o', 4) '+', 5) 'x' .

```python
plot(w, abs(fft(x)), label='200')
plot(w1, abs(fft(x[:4])), 'yo', label='4')
plot(w2, abs(fft(x[:8])), 'g+', label='8')
plot(w3, abs(fft(x[:16])), 'rx', label='16')
legend()
```

7) Comente o resultado.

Quando adicionamos mais zeros no sinal discreto, a transformada terá mais pontos intermediários e poderemos vizualizar melhor o comportamento da curva.

8) Gere um vetor x1, uma senoide de 32 pontos, com 2 ciclos completos.

```python
n = arange(0,32,1)
x1 = sin(n*2*2*pi/32)
ws1 = linspace(0, 2*pi, 32, endpoint=False)
stem(ws1, x1)
```

9) Gere vetor x2 acrescentando zeros ao final do vetor x de forma a duplicar seu tamanho.


```python
x2 = zeros(64)
x2[:32] = x1
ws2 = linspace(0, 2*pi, 64, endpoint=False)
stem(ws2, x2)
```

10) Gere um vetor x3 com tamanho 128 acrescentando mais zeros


```python
x3 = zeros(128)
x3[:32] = x1
ws3 = linspace(0, 2*pi, 128, endpoint=False)
stem(ws3, x3)
```

11) Plote no mesmo eixo de frequencia (de 0 a 2pi) o valor absoluto das DFTs X1, X2 e X3 e comente o resultado.
(simbolos X1: 'o', X2: '+', X3: '-'

```python
plot(ws1, abs(fft(x1)), 'bo', label='32')
plot(ws2, abs(fft(x2)), 'r+', label='64')
plot(ws3, abs(fft(x3)), 'g-', label='128')
legend()
```
