```python
from pylab import *
```

```python
# função que plota o sinal x e sua TF em função de n
def plota(n,x):
    X=fft(x)/len(x)
    figure(figsize=(15,4))
    subplot(1,2,1)
    plot(n,x)
    subplot(1,2,2)
    stem(n,abs(X),linefmt='C0-',markerfmt='C0o',basefmt='C0-')
    show()
    ## informa o valor máximo da energia e em que posição se encontra
    print("Posição: ", argmax(X))
    print("Energia: ", abs(X[argmax(X)]))
```

I) Implemente função que plota um número (inteiro ou não de ciclos de cossenoide)

```python
def plotaciclos(C):
    N=100
    n=arange(N)
    W=2*pi/N
    w=W*C
    x=cos(w*n)
    plota(n,x)
```
