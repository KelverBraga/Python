1) Implementar uma função X=tfourier(x,N) que calcule N valores da transformada de fourier da sequencia representada pelo *array* x. Ou seja, o *array* X vai conter N elementos, cada um correspondente ao valor transformado para a frequencia $w$, (0 a $2\pi$).

```python
def tfourier(x, N):
    """retorna a transformada de Fourier de x para N valores de w entre 0 e 2pi"""
    ##utilizando 2 for
    #X = zeros(N, dtype = complex)
    #for k in range(N):
    #    for n in range(len(x)):
    #        X[k] += x[n] * exp(-1j * 2 * pi * (k / N) * n)
    
    ##utilizando matrizes
    col = arange(N).reshape(N,1)            #cria um array coluna de N elementos
    lin = arange(len(x)).reshape(1,len(x))  #cria um array linha com o tamanho de x

    w = exp(-1j*2*pi* (col @ lin) /N)         #cria a matriz complexa

    X = w @ x
    
    return X
#tfourier=fft
```

```python
# a)
N=100
x1=concatenate((zeros(10),ones(10),zeros(10)))

X1t=tfourier(x1,N)
X1f=fft(x1)
#print(shape(X1t))
#print(shape(X1f))
```

```python
figure(figsize=(15,4))
subplot(1,2,1)
plot(x1,label='x1')
xlabel('n')
legend()
subplot(1,2,2)
xlabel('w (rad)')

k=2*pi*arange(N)/N
kk=2*pi*arange(len(X1f))/len(X1f)
plot(k, abs(X1t),label='tfourier')
plot(kk, abs(X1f),'o',label='fft')
legend()
show()
```
